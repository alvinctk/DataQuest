## Reading in Data

````python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt 
from matplotlib.ticker import MaxNLocator
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
````

````python
star_wars.head()
````

````python
print(star_wars.columns)
````

## Cleaning and Mapping Yes/No Columns


````python
yes_no = {"Yes":True, "No":False, np.nan:False}
columns = ["Have you seen any of the 6 films in the Star Wars franchise?",
           "Do you consider yourself to be a fan of the Star Wars film franchise?"]

for col in columns:
    star_wars[col] = star_wars[col].map(yes_no)

star_wars.head()
    
````

## Cleaning and Mapping Checkbox Columns
Using [pandas.DataFrame.rename()](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.rename.html) to rename columns to reflect what they represent.
<code><font color="red">df.rename()</font></code>, works a lot like a <code><font color="red">map()</font></code>, with a dictionary argument that maps current columns names to new one.

Changing <code><font color="red">Which of the following Star Wars films have you seen? Please select all that apply.</font></code> column to <code><font color="red">seen_1</font></code>.

````python
print(star_wars.columns[3:9])


movie_mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_mapping)
    
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
    
})

star_wars[3:9].head()
````

````python
rename_map = dict(zip(star_wars.columns[9:15].tolist(), ["ranking_" + str(x-8) for x in range(9,15)]))
print(rename_map)
star_wars = star_wars.rename(columns=rename_map)
````

## Finding the Highest-Ranked Movie
Using [pandas.DataFrame.mean()](http://pandas.pydata.org/pandas-docs/version/0.17.1/generated/pandas.DataFrame.mean.html)

````python
%matplotlib inline

# Use actual title in display since ranking_x has no contextual meaning  
movies = {"ranking_1": "Star Wars: Episode I The Phantom Menace", 
"ranking_2": "Star Wars: Episode II Attack of the Clones", 
"ranking_3": "Star Wars: Episode III Revenge of the Sith",
"ranking_4": "Star Wars: Episode IV A New Hope",
"ranking_5": "Star Wars: Episode V The Empire Strikes Back",
"ranking_6": "Star Wars: Episode VI Return of the Jedi"}

# Data to bar plot, we want the least value to the be first 
mean = star_wars[star_wars.columns[9:15]].mean().sort_values(ascending=False)

# To determine which movie is Best/Worst ranked
best, worst = min(mean), max(mean)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18,7))
fig.canvas.set_window_title('Star Wars Ranked Movies')
fig.subplots_adjust(left=0.115, right=0.88)

# Remove all borders (spines) of the plot -> result in white grid
for edge, spine in ax.spines.items():
    spine.set_visible(False)
    
for y, x in mean.iteritems():
    y = movies[y]
    color = "grey"
    kw = dict(textcoords="offset points", fontsize=16)
    if x == best:
        color = "blue"
        kw["weight"] ="bold"
        
        # Arrow to indicate the best ranked
        ax.annotate("Best Ranked",
             xy=(x, y),  xycoords='data',
             xytext=(+30, +10), **kw,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    elif x == worst:
        
        # Arrow to indicate the worst ranked
        ax.annotate("Worst Ranked", xy=(x, y), xycoords='data', xytext=(+30, +10), **kw,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
        data_source = "Data source: FiveThirtyEight;"
        data_source += " Includes " + str(len(star_wars)) + " number of survey respondents."
        data_source += "Note that repsondent were able to choose up to six options."
        ax.annotate(data_source, xy=(x, y),  xycoords='data', xytext=(-180, -50), textcoords="offset points", fontsize=20, horizontalalignment='center', size='medium', transform=ax.transAxes)

         
    # If you are using rect method. But in this example, we are not.     
    rect,  = ax.barh(y, x, color=color)
    
    # Annotate text color 
    kw["color"] ="white"
    
    # Annotate the mean values for each horiztonal bar 
    ax.annotate("{:.2f}".format(x), xy=(x, y), xycoords="data", xytext=(-48, 0), **kw)
    
    # Annotate Movies for each horiztonal bar
    ax.annotate(y, xy=(0, y), xycoords="data", xytext=(+20, 0), **kw)
    
# Remove ticks and labels on both axis and major/minor ticks   
plt.tick_params(
    axis="both",       # changes apply to both x-axis and y-axis
    which="both",      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    right=False,       # ticks along the right edge are off
    left=False,        # ticks along the left edge are off
    labelbottom=False, # labels along the bottom edge are off
    labeltop=False,    # labels along the top edge are off
    labelright=False,  # labels along the right edge are off
    labelleft=False    # labels along the left edge are off
) 

ax.set_title("Star Wars Survey Ranked Movies According To Mean", fontsize=20, pad=20)
plt.savefig("barh_star_wars_survey_ranked_movies.png",dpi=72)
plt.show()
````

We observed "Star Wars: Episode V The Empire Strikes Back" is the best ranked movie. As I have suspected, the original movies are rate much highly than the new movies.

````python
%matplotlib inline

# Use actual title in display since ranking_x has no contextual meaning  
seen_movies = {"seen_1": "Star Wars: Episode I The Phantom Menace", 
"seen_2": "Star Wars: Episode II Attack of the Clones", 
"seen_3": "Star Wars: Episode III Revenge of the Sith",
"seen_4": "Star Wars: Episode IV A New Hope",
"seen_5": "Star Wars: Episode V The Empire Strikes Back",
"seen_6": "Star Wars: Episode VI Return of the Jedi"}

# Data to bar plot -> We want the top bar to be most viewed movie
seen_sum = star_wars[star_wars.columns[3:9]].sum().sort_values(ascending=True)

# To determine which movie is Most/Least viewed
best, worst = max(seen_sum), min(seen_sum)


fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18,7))
fig.canvas.set_window_title('Star Wars Viewed Movies')
fig.subplots_adjust(left=0.115, right=0.88)

# Remove all borders (spines) of the plot -> result in white grid
for edge, spine in ax.spines.items():
    spine.set_visible(False)

for y, x in seen_sum.iteritems():
    
    y = seen_movies[y]
    
    color = "grey"
    kw = dict(textcoords="offset points", fontsize=16)
    if x == best:
        color = "blue"
        kw["weight"] ="bold"
        
        # Arrow to indicate the best ranked
        ax.annotate("Most Viewed",
             xy=(x, y),  xycoords='data',
             xytext=(+30, +10), **kw,
             arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
    
    elif x == worst:
        
        # Arrow to indicate the worst ranked
        ax.annotate("Least Viewed", xy=(x, y), xycoords='data', xytext=(+30, +10), **kw,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
        data_source = "Data source: FiveThirtyEight;"
        data_source += " Includes " + str(len(star_wars)) + " number of survey respondents."
        data_source += "Note that repsondent were able to choose up to six options."
        ax.annotate(data_source, xy=(x, y),  xycoords='data', xytext=(-180, -50), textcoords="offset points", fontsize=20, horizontalalignment='center', size='medium', transform=ax.transAxes)

         
    # If you are using rect method. But in this example, we are not.     
    rect,  = ax.barh(y, x, color=color)
    
    # Annotate text color 
    kw["color"] ="white"
    
    # Annotate the mean values for each horiztonal bar 
    ax.annotate("{:.2f}".format(x), xy=(x, y), xycoords="data", xytext=(-66, 0), **kw)
    
    # Annotate Movies for each horiztonal bar
    ax.annotate(y, xy=(0, y), xycoords="data", xytext=(+20, 0), **kw)
    
# Remove ticks and labels on both axis and major/minor ticks   
plt.tick_params(
    axis="both",       # changes apply to both x-axis and y-axis
    which="both",      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    right=False,       # ticks along the right edge are off
    left=False,        # ticks along the left edge are off
    labelbottom=False, # labels along the bottom edge are off
    labeltop=False,    # labels along the top edge are off
    labelright=False,  # labels along the right edge are off
    labelleft=False    # labels along the left edge are off
) 

ax.set_title("Star Wars Survey Most Viewed Movie", fontsize=20, pad=20)
plt.savefig("barh_star_wars_survey_viewed_movies.png",dpi=72)
plt.show()
````

Original movies were seen more by respondents than the new movies. This reinforces the findings that original movies are more popular.

## Exploring the Data by Binary Segments
We are interested in exploring data in gender segements. 

````python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
````

````python
# Use actual title in display since ranking_x has no contextual meaning  
seen_movies = {"seen_1": "Star Wars: Episode I The Phantom Menace", 
"seen_2": "Star Wars: Episode II Attack of the Clones", 
"seen_3": "Star Wars: Episode III Revenge of the Sith",
"seen_4": "Star Wars: Episode IV A New Hope",
"seen_5": "Star Wars: Episode V The Empire Strikes Back",
"seen_6": "Star Wars: Episode VI Return of the Jedi"}

movies = {"ranking_1": "Star Wars: Episode I The Phantom Menace", 
"ranking_2": "Star Wars: Episode II Attack of the Clones", 
"ranking_3": "Star Wars: Episode III Revenge of the Sith",
"ranking_4": "Star Wars: Episode IV A New Hope",
"ranking_5": "Star Wars: Episode V The Empire Strikes Back",
"ranking_6": "Star Wars: Episode VI Return of the Jedi"}


def plot_best_barhs(data, mapping, best, worst, title, saved, str_format, x_annotate):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(18,7))
    fig.canvas.set_window_title('Star Wars Viewed Movies')
    fig.subplots_adjust(left=0.115, right=0.88)
    
    # Remove all borders (spines) of the plot -> result in white grid
    for edge, spine in ax.spines.items():
        spine.set_visible(False)
        
    for y, x in data.iteritems():
        y = mapping[y]
        color = "grey"
        kw = dict(textcoords="offset points", fontsize=16)
        if x == best:
            color = "blue"
            kw["weight"] ="bold"
            # Arrow to indicate the best ranked
            ax.annotate("Most Viewed",xy=(x, y),  xycoords='data', xytext=(+30, +10), **kw,
                        arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
        elif x == worst:
            # Arrow to indicate the worst ranked
            ax.annotate("Least Viewed", xy=(x, y), xycoords='data', xytext=(+30, +10), **kw,arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
                
            data_source = "Data source: FiveThirtyEight;"
            data_source += " Includes " + str(len(star_wars)) + " number of survey respondents."
            data_source += "Note that repsondent were able to choose up to six options."
            ax.annotate(data_source, xy=(x, y),  xycoords='data', xytext=(-180, -50), textcoords="offset points", fontsize=20, horizontalalignment='center', size='medium', transform=ax.transAxes)

         
        # If you are using rect method. But in this example, we are not.     
        rect,  = ax.barh(y, x, color=color)
    
        # Annotate text color 
        kw["color"] ="white"
    
        # Annotate the mean values for each horiztonal bar 
        ax.annotate(str_format.format(x), xy=(x, y), xycoords="data", xytext=(x_annotate, 0), **kw)
    
        # Annotate Movies for each horiztonal bar
        ax.annotate(y, xy=(0, y), xycoords="data", xytext=(+20, 0), **kw)
    
    # Remove ticks and labels on both axis and major/minor ticks   
    plt.tick_params(
        axis="both",       # changes apply to both x-axis and y-axis
        which="both",      # both major and minor ticks are affected
        bottom=False,      # ticks along the bottom edge are off
        top=False,         # ticks along the top edge are off
        right=False,       # ticks along the right edge are off
        left=False,        # ticks along the left edge are off
        labelbottom=False, # labels along the bottom edge are off
        labeltop=False,    # labels along the top edge are off
        labelright=False,  # labels along the right edge are off
        labelleft=False    # labels along the left edge are off
    ) 

    ax.set_title(title, fontsize=20, pad=20)
    plt.savefig(saved,dpi=72)
    plt.show()
````

## Male/Female differencese in favorite Star Wars movie and most seen movies.

````python
males_ranked_movies = males[star_wars.columns[9:15]].mean().sort_values(ascending=False)
females_ranked_movies = females[star_wars.columns[9:15]].mean().sort_values(ascending=False)


plot_best_barhs(males_ranked_movies, movies, min(males_ranked_movies), max(males_ranked_movies), 
                "Star Wars Survey Males Ranked Movies", "Star-Wars-Survey-Males-Ranked-Movies.png", 
                "{:.2f}", -50)

plot_best_barhs(females_ranked_movies, movies, min(females_ranked_movies), max(females_ranked_movies), 
                "Star Wars Survey Females Ranked Movies", "Star-Wars-Survey-Females-Ranked-Movies.png",
               "{:.2f}", -50)
````

````python
males_viewed_movies = males[star_wars.columns[3:9]].sum().sort_values(ascending=True)
females_viewed_movies = females[star_wars.columns[3:9]].sum().sort_values(ascending=True)
plot_best_barhs(males_viewed_movies, seen_movies, max(males_viewed_movies), min(males_viewed_movies), 
                "Star Wars Survey Males Viewed Movies", "Star-Wars-Survey-Males-Viewed-Movies.png",
               "{:d}", -40)

plot_best_barhs(females_viewed_movies, seen_movies, max(females_viewed_movies), min(females_viewed_movies), 
                "Star Wars Survey Females Viewed Movies", "Star-Wars-Survey-Females-Viewed-Movies.png",
               "{:d}", -40)
````

More males watch episodes 1-3, but males liked them far less than females did.
